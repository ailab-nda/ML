##　モデルのロード

```python
import gensim

MODEL_PATH = "model.vec"
model = gensim.models.KeyedVectors.load_word2vec_format(MODEL_PATH, binary=False)
```

## 実行

### 単語 --> ベクトルの値

```python
model.get_vector('男')
```

### ベクトルが似ている単語の検索

```python
for item, value in model.most_similar('人生'):
    print(item,value)
```

### ベクトルによる単語の演算

```python
print(model.most_similar(positive=['国王', '女'], negative=['男']))
```
